```python
def get_all_images(paths):
    src = []
    for path in paths:
        df = pd.read_csv(path)

        temp_src = df.loc[:, "src"].to_list()       
        src = src + temp_src

    print(src)
```

```python
get_all_images(all_datas)
```

```python
def make_indexing(path):
    df = pd.read_csv(path)

    temp_data = df.loc[:, ["id", "src", "caption"]]
    ids = temp_data["id"].to_list()
    temp_data = temp_data.set_index("id")
    dict_data = temp_data.to_dict()
    
    result = {_id:{"src":dict_data["src"][_id], "caption":dict_data["caption"][_id]} for _id in ids}

    return result

def make_index_table(paths):
    result = {}

    for path in paths: result.update(make_indexing(path))

    return result
```

```python
index_table = make_index_table(all_datas)
```

```python
import json

with open("dataest_index_table.json", "a", encoding="utf-8") as f:
    json.dump(index_table, f)
```
