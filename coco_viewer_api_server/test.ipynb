```python
true_labels = {i:[] for i in list(range(10))}
true_labels
```

```python
images = glob.glob("/Users/yoonseonghyeon/Downloads/mnist_png/training/*/*")
for image in images:
    label = int(image.split("/")[-2])
    true_src = "/"+"/".join(image.split("/")[-4:])
    true_labels[label].append(true_src)
```

```python
true_labels
```

```python
data = {"images": [], "annotations": []}
idx = 0
for key in true_labels.keys():
    items = true_labels[key]
    for item in items:
        idx += 1
        id = item.split("/")[-1].split(".")[0]
        data["images"].append({"file_name": item, "id": id, "width": 28, "height": 28})
        data["annotations"].append({"id": idx, "image_id": id, "caption": key})
```

```python
import json
with open("dataset.json", "w") as f:
    json.dump(data, f)
```
