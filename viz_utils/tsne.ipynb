```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = load_digits()
```

```python
data.data.shape
```

```python
data.target
```

```python
n_dim = 2
model = TSNE(n_components=n_dim)

X_embedded = model.fit_transform(data.data)
```

```python
palette = sns.color_palette("bright", 10)
# plt.figure(dpi=1200
sns.scatterplot(x=X_embedded[:, 0], y=X_embedded[:, 1], hue="target", data=data, alpha=0.6)
plt.axis("off")
plt.legend().set_visible(False)
plt.show()
```
